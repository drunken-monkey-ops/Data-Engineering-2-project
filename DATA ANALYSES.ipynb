# Question: 2

Q2. What are the top 10 most frequently updated GitHub projects (i.e., most commits in a
project)?

```python
def truncate_text(text, max_length=15):
    return text if len(text) <= max_length else text[:max_length] + '...'
all_documents = collection.find()  # Re-fetch the documents
documents_list = list(all_documents)
df = pd.DataFrame(documents_list)
df['commits'] = pd.to_numeric(df['commits'], errors='coerce')
top_projects_by_commits = df.sort_values(by='commits', ascending=False).head(10)
top_projects_by_commits['name'] = top_projects_by_commits['name'].apply(truncate_text)

print("Top 10 Most Frequently Updated GitHub Projects (by number of commits):")
print(top_projects_by_commits[['name', 'owner', 'commits']])


plt.figure(figsize=(12, 8))
plt.bar(top_projects_by_commits['name'], top_projects_by_commits['commits'], color='skyblue')
plt.xlabel('GitHub Projects')
plt.ylabel('Number of Commits')
plt.title('Top 10 Most Frequently Updated GitHub Projects')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Q.2.Top 10 Most Frequently Updated GitHub Projects.png', format='png')
plt.show()
```

## Q3 
Q3. What are the top 10 programming languages that follow the test-driven development
approach (i.e., most projects with unit tests)?

```python
all_documents = collection.find()  # Re-fetch the documents
documents_list = list(all_documents)
df = pd.DataFrame(documents_list)

df['languages'] = df['languages'].apply(lambda x: x if isinstance(x, list) else [])
df['has_tests'] = df['has_tests'].astype(bool)

df_with_tests = df[df['has_tests'] == True]
df_exploded = df_with_tests.explode('languages')

language_counts = df_exploded['languages'].value_counts().reset_index()
language_counts.columns = ['language', 'count']

top_languages_with_tests = language_counts.head(10)
print("Top 10 Programming Languages that Follow the Test-Driven Development Approach:")
print(top_languages_with_tests)

plt.figure(figsize=(12, 8))
plt.bar(top_languages_with_tests['language'], top_languages_with_tests['count'], color='skyblue')
plt.xlabel('Programming Languages')
plt.ylabel('Number of Projects with Unit Tests')
plt.title('Top 10 Programming Languages with Test-Driven Development')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Q.3.Top 10 Programming Languages with Test-Driven Development.png', format='png')
plt.show()
```

```python
all_documents = collection.find()  # Re-fetch the documents
documents_list = list(all_documents)
df = pd.DataFrame(documents_list)

df['languages'] = df['languages'].apply(lambda x: x if isinstance(x, list) else [])
df['has_tests'] = df['has_tests'].astype(bool)
df['has_ci_cd'] = df['has_ci_cd'].astype(bool)


df_with_tests_and_ci = df[(df['has_tests'] == True) & (df['has_ci_cd'] == True)]
df_exploded = df_with_tests_and_ci.explode('languages')

language_counts = df_exploded['languages'].value_counts().reset_index()
language_counts.columns = ['language', 'count']
top_languages_with_tests_and_ci = language_counts.head(10)

print("Top 10 Programming Languages that Follow TDD and DevOps Approach:")
print(top_languages_with_tests_and_ci)


plt.figure(figsize=(12, 8))
plt.bar(top_languages_with_tests_and_ci['language'], top_languages_with_tests_and_ci['count'], color='skyblue')
plt.xlabel('Programming Languages')
plt.ylabel('Number of Projects with Unit Tests and CI/CD')
plt.title('Top 10 Programming Languages with TDD and DevOps Approach')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Q.4.Top 10 Programming Languages with TDD and DevOps Approach.png', format='png')
plt.show()
```
