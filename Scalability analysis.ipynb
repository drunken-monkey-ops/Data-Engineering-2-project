```python
import matplotlib.pyplot as plt

# Provided execution times (in seconds)
execution_times = {
    1: 13208.323948860168,
    2: 6956.650584459305,
    3: 5247.4493935108185
}

# Calculate speedups based on the execution times
speedup = {n: execution_times[1] / t for n, t in execution_times.items()}

# Calculate efficiencies based on the speedups
efficiency = {n: s / n for n, s in speedup.items()}

# Extract the number of nodes for plotting
nodes = sorted(execution_times.keys())

# Prepare data for plotting
speedup_values = [speedup[n] for n in nodes]
efficiency_values = [efficiency[n] for n in nodes]
ideal_speedup = nodes
ideal_efficiency = [1 for _ in nodes]

# Define figure size
figsize = (10, 6)

# Plotting the Speedup Strong Scaling
plt.figure(figsize=figsize)

plt.plot(nodes, speedup_values, marker='o', label='Measured Speedup')
plt.plot(nodes, ideal_speedup, linestyle='--', label='Ideal Speedup')
plt.xlabel('Number of Consumers')
plt.ylabel('Speedup')
plt.title('Speedup Strong Scaling')
plt.xticks(nodes)  # Only show integers 1, 2, 3
plt.legend()
plt.grid(True)

plt.savefig('Speedup_Strong_Scaling_111.png', format='png')
plt.show()

# Plotting the Efficiency Strong Scaling
plt.figure(figsize=figsize)

plt.plot(nodes, efficiency_values, marker='o', label='Measured Efficiency')
plt.plot(nodes, ideal_efficiency, linestyle='--', label='Ideal Efficiency')
plt.xlabel('Number of Consumers')
plt.ylabel('Efficiency')
plt.ylim(0,)
plt.title('Efficiency Strong Scaling')
plt.xticks(nodes)  # Only show integers 1, 2, 3
plt.legend()
plt.grid(True)

plt.savefig('Efficiency_Strong_Scaling_1111.png', format='png')
plt.show()
```
